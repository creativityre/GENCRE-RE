```python
import pandas as pd

# Load CSV file
df = pd.read_csv("novelty.csv")

# Columns to check
columns = ["RQ1", "RQ2", "RQ3", "RQ4", "RQ5"]

# Copy dataframe
clean_df = df.copy()

# Remove outliers using IQR method
for col in columns:
    Q1 = clean_df[col].quantile(0.25)
    Q3 = clean_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    clean_df = clean_df[
        (clean_df[col] >= lower_bound) &
        (clean_df[col] <= upper_bound)
    ]

# Save cleaned data
clean_df.to_csv("cleaned_data_got_novelty.csv", index=False)

print("Original shape:", df.shape)
print("Cleaned shape:", clean_df.shape)

print("\nCleaned Data:")
print(clean_df)
```

```python
import pandas as pd
import pingouin as pg

df = pd.read_csv("cleaned_data_got_novelty.csv")

# assume columns = Req1..Req5, rows = raters
df_long = df.reset_index().melt(id_vars="index")
df_long.columns = ["rater", "requirement", "score"]

icc = pg.intraclass_corr(
    data=df_long,
    targets="requirement",
    raters="rater",
    ratings="score"
)

print(icc)
```
