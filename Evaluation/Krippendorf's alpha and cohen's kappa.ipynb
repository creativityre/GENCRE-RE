```python
import pandas as pd
import numpy as np
from statsmodels.stats.inter_rater import fleiss_kappa

df = pd.read_csv("1504 GPT.csv", encoding=result['encoding'])

count_matrix = []

for _, row in df.iterrows():
    ratings = row[['Reviewer 1', 'Reviewer 2', 'Reviewer 3']]

    count_0 = (ratings == 0).sum()
    count_1 = (ratings == 1).sum()

    count_matrix.append([count_0, count_1])

count_matrix = np.array(count_matrix)

kappa = fleiss_kappa(count_matrix)

print(f"Fleiss' Kappa = {kappa:.4f}")
```

```python
import pandas as pd

df_gpt = pd.read_csv("1504 GPT.csv", encoding="ISO-8859-1")
df_gemini = pd.read_csv("2126 Gemini.csv", encoding="ISO-8859-1")
```

```python
df_gpt["requirement"] = "GPT_" + df_gpt["ieee_requirement"].astype(str)
df_gemini["requirement"] = "GEMINI_" + df_gemini["ieee_requirement"].astype(str)
```

```python
gpt_long = df_gpt.melt(
    id_vars=["requirement"],
    value_vars=["Reviewer 1", "Reviewer 2", 'Reviewer 3'],
    var_name="reviewer",
    value_name="rating"
)

gpt_long["system"] = "GPT"
```

```python
gemini_long = df_gemini.melt(
    id_vars=["requirement"],
    value_vars=["Reviewer 1", "Reviewer 2", 'Reviewer 3'],
    var_name="reviewer",
    value_name="rating"
)

gemini_long["system"] = "Gemini"
```

```python
final_df = pd.concat([gpt_long, gemini_long], ignore_index=True)
```

```python
final_df["reviewer"] = final_df["reviewer"].str.replace("Reviewer ", "R")
```

```python
final_df.to_csv("combined_mixed_effects_data_1.csv", index=False)
```

```python
pip install bambi pymc
```

```python
import bambi as bmb

model = bmb.Model(
    "rating ~ system + (1|system:requirement) + (1|reviewer)",
    final_df,
    family="bernoulli"
)

result = model.fit()
print(result)
```
